# Plots

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is widely used for generating plots, charts, and graphs, making it an essential tool for data visualization. Matplotlib is highly customizable and supports a variety of plot types, including:

* Line plots
* Scatter plots
* Bar charts
* Histograms
* Pie charts
* 3D plots

Matplotlib integrates well with other libraries such as NumPy and Pandas, allowing for seamless data manipulation and visualization. It is also compatible with Jupyter Notebooks, making it a popular choice for data scientists and analysts.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a plot
plt.plot(x, y, label='Sine Wave', color='blue', linestyle='--')

# Add grid
plt.grid(True)

# Add legend
plt.legend()

# Add title and labels
plt.title('Customized Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()

# Save the plot
plt.savefig('sine_wave.png')
```

```python
# Data for line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Data for scatter plot
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

# Data for bar chart
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 18]

# Data for histogram
data = np.random.randn(1000)

# Data for pie chart
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
```

```python
plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
plt.scatter(x_scatter, y_scatter, color='red')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
plt.bar(categories, values, color='green')
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
plt.hist(data, bins=30, color='purple')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart')
plt.show()
```

```python
plt.plot(x, y)
plt.grid(True)
plt.title('Line Plot with Grid')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
plt.plot(x, y, label='Sine Wave')
plt.legend()
plt.title('Line Plot with Legend')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line plot
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Line Plot')

# Scatter plot
axs[0, 1].scatter(x_scatter, y_scatter, color='red')
axs[0, 1].set_title('Scatter Plot')

# Bar chart
axs[1, 0].bar(categories, values, color='green')
axs[1, 0].set_title('Bar Chart')

# Histogram
axs[1, 1].hist(data, bins=30, color='purple')
axs[1, 1].set_title('Histogram')

# Adjust layout
plt.tight_layout()
plt.show()
```
